```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('azodata.csv')
```

```python
data.isnull().sum()
data.dropna(inplace=True)
```

```python
data_dict = {
    'date': ['january', 'february', 'march', 'april', 'may', 'june',
             'july', 'august', 'september', 'october', 'november', 'december'],
    'season': ['winter', 'winter', 'spring', 'spring', 'spring', 'summer',
               'summer', 'summer', 'fall', 'fall', 'fall', 'winter'],
    'step_num_avg': [8247, 9280, 12847, 13722, 23718, 18379,
                     15755, 14797, 15318, 28592, 22489, 17239]
}


df = pd.DataFrame(data_dict)


df['month_index'] = pd.to_datetime(df['date'], format='%B').dt.month


df_sorted = df.sort_values('month_index')


plt.figure(figsize=(12, 6))
plt.plot(df_sorted['date'], df_sorted['step_num_avg'], marker='o', color='blue')
plt.title('Average Monthly Step Count by Season 3 Years')
plt.xlabel('Month')
plt.ylabel('Average Step Count By Three Years')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

data_dict = {
    'date': ['january', 'february', 'march', 'april', 'may', 'june',
             'july', 'august', 'september', 'october', 'november', 'december'],
    'season': ['winter', 'winter', 'spring', 'spring', 'spring', 'summer',
               'summer', 'summer', 'fall', 'fall', 'fall', 'winter'],
    'step_num_avg': [8247, 9280, 12847, 13722, 23718, 18379,
                     15755, 14797, 15318, 28592, 22489, 17239]
}


df = pd.DataFrame(data_dict)

df['month_index'] = pd.to_datetime(df['date'], format='%B').dt.month

df_sorted = df.sort_values('month_index')

plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
plt.hist(df_sorted['step_num_avg'], color='skyblue', bins=12)
plt.title('Step Count Distribution')
plt.xlabel('Average Step Count')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.scatter(df_sorted['month_index'], df_sorted['step_num_avg'], color='green')
plt.title('Average Step Count by Month')
plt.xlabel('Month')
plt.ylabel('Average Step Count')
plt.grid(True)
plt.xticks(df_sorted['month_index'], df_sorted['date'], rotation=45)

plt.tight_layout()
plt.show()
```
